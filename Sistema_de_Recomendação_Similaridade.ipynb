```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

from unicodedata import normalize
from string import punctuation # Retirar todas as pontuações
```

```python
nltk.download('stopwords')
nltk.download('punkt')
```

```python
stop_words = stopwords.words('english')
```

```python
def remover_sw(texto):
  words_texto = word_tokenize(texto)
  word_sem_sw = [word for word in words_texto if not word in stop_words]
  return ' '.join(word_sem_sw)
```

```python
def limparTexto(texto):
  # Remover caracteres especiais
  texto = normalize('NFKD', texto).encode('ASCII','ignore').decode('utf-8')
  #Remover ponuações
  texto = re.sub(f'[{punctuation}]','', texto)
  #Remover números
  texto = re.sub(r'[0-9]','', texto)
  # Remover caracteres japoneses
  texto = re.sub(r'[ぁ-ゟ]','', texto)
  #Colocar em caixa baixa
  texto  = texto.lower()
  #Remover stopwords
  texto = remover_sw(texto)
  return texto
```

```python
nltk.download('punkt_tab')
```

```python
%%time
df['synopsis_limpo'] = df['synopsis'].map(limparTexto)
```

# PLN - Recomendação com foco na sinopse dos animes

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer(analyzer='word', ngram_range = (1,1))
```

```python
vetores = vectorizer.fit_transform(df['synopsis_limpo'])
```

```python
print(vetores)
```

```python
# Apenas para saber o que tem dentro da matriz

words_df = pd.DataFrame(vetores.toarray(),
                        columns = vectorizer.get_feature_names_out())
```

```python
words_df
```

```python
# Construiremos a base para calcular as recomendações
%%time
cossenos_sinopse = cosine_similarity(words_df, words_df)
```

```python
def recomendar(nome, matriz_similaridades):
  # Obter o índice com base no nome informado
  indice = indices[nome]

  # Obter similaridades para o item selecionado
  similaridades = (enumerate(matriz_similaridades[indice]))

  similaridades = sorted(similaridades, key=lambda x: x[1], reverse=True)
  # Obter os 10 mais similares
  similaridades = similaridades[1:11]

  #Obter o indices dos similares
  animes_indices = [x[0]for x in similaridades]
  animes_similaridade = [x[1] for x in similaridades]

  for i,s in zip (animes_indices, animes_similaridade):
    print(df.loc[i,'name'], '- com similaridade igual a ', s)
```

```python
recomendar('Drifters', cossenos_sinopse)
```
